```python
from neuralprophet import NeuralProphet
import pandas as pd

# Load data
df = pd.read_csv("combined.csv")

# Prepare for NeuralProphet
df_np = df.groupby("year", as_index=False)["Total_cases"].mean()
df_np.columns = ["ds", "y"]
df_np["ds"] = pd.to_datetime(df_np["ds"], format="%Y")

# Train model
m = NeuralProphet(yearly_seasonality=True)
m.fit(df_np, freq="Y")

# Forecast
future = m.make_future_dataframe(df_np, periods=5)
forecast = m.predict(future)

# Save forecast
out = forecast[["ds","yhat1"]]
out["year"] = out["ds"].dt.year
out.rename(columns={"yhat1":"np_forecast"}, inplace=True)

out[["year","np_forecast"]].to_csv("neuralprophet_forecast.csv", index=False)
```
